```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Iris dataset
iris_df = pd.read_csv("/content/Iris.csv")

# Set the style of seaborn
sns.set_style("whitegrid")

# Univariate visualizations for each feature
plt.figure(figsize=(20, 15))
1
# Plot 1: Scatter plot for Sepal Length vs Sepal Width
plt.subplot(3, 4, 1)
sns.scatterplot(data=iris_df, x="SepalLengthCm", y="SepalWidthCm", hue="Species", palette="viridis")
plt.title("Sepal Length vs Sepal Width")
```

```python
# Set the style of seaborn
sns.set_style("whitegrid")

# Univariate visualizations for each feature
plt.figure(figsize=(20, 15))
# Plot 2: Scatter plot for Petal Length vs Petal Width
plt.subplot(3, 4, 2)
sns.scatterplot(data=iris_df, x="PetalLengthCm", y="PetalWidthCm", hue="Species", palette="viridis")
plt.title("Petal Length vs Petal Width")
```

```python
# Set the style of seaborn
sns.set_style("whitegrid")

# Univariate visualizations for each feature
plt.figure(figsize=(20, 15))
# Plot 3: Bar plot for Sepal Length
plt.subplot(3, 4, 3)
sns.barplot(data=iris_df, x="Species", y="SepalLengthCm", palette="viridis")
plt.title("Sepal Length by Species")
```

```python
# Set the style of seaborn
sns.set_style("whitegrid")

# Univariate visualizations for each feature
plt.figure(figsize=(20, 15))
# Plot 4: Bar plot for Sepal Width
plt.subplot(3, 4, 4)
sns.barplot(data=iris_df, x="Species", y="SepalWidthCm", palette="viridis")
plt.title("Sepal Width by Species")
```

```python
# Set the style of seaborn
sns.set_style("whitegrid")

# Univariate visualizations for each feature
plt.figure(figsize=(20, 15))
# Plot 5: Boxplot for Sepal Length
plt.subplot(3, 4, 5)
sns.boxplot(data=iris_df, x="Species", y="SepalLengthCm", palette="viridis")
plt.title("Sepal Length Boxplot")
```

```python
# Plot 6: Boxplot for Sepal Width
plt.subplot(3, 4, 6)
sns.boxplot(data=iris_df, x="Species", y="SepalWidthCm", palette="viridis")
plt.title("Sepal Width Boxplot")
```

```python
# Set the style of seaborn
sns.set_style("whitegrid")

# Univariate visualizations for each feature
plt.figure(figsize=(20, 15))

# Plot 7: Histogram for Sepal Length
plt.subplot(3, 4, 7)
sns.histplot(data=iris_df, x="SepalLengthCm", kde=True, color='skyblue')
plt.title("Sepal Length Histogram")

# Plot 8: Histogram for Sepal Width
plt.subplot(3, 4, 8)
sns.histplot(data=iris_df, x="SepalWidthCm", kde=True, color='salmon')
plt.title("Sepal Width Histogram")

# Plot 9: Histogram for Petal Length
plt.subplot(3, 4, 9)
sns.histplot(data=iris_df, x="PetalLengthCm", kde=True, color='lightgreen')
plt.title("Petal Length Histogram")

# Plot 10: Histogram for Petal Width
plt.subplot(3, 4, 10)
sns.histplot(data=iris_df, x="PetalWidthCm", kde=True, color='gold')
plt.title("Petal Width Histogram")
```
