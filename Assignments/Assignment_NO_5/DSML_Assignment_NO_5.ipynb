```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Iris dataset
iris_df = pd.read_csv("/content/Iris.csv")

# Set the style of seaborn
sns.set_style("whitegrid")

# Univariate visualizations for each feature
plt.figure(figsize=(15, 10))
```

```python
print(iris_df)
```

```python

# Plot 1: Sepal Length Distribution
plt.subplot(2, 3, 1)
sns.histplot(iris_df["SepalLengthCm"], kde=True, color='skyblue')
plt.title("Sepal Length Distribution")
```

```python

# Plot 2: Sepal Width Distribution
plt.subplot(2, 3, 2)
sns.histplot(iris_df["SepalWidthCm"], kde=True, color='salmon')
plt.title("Sepal Width Distribution")
```

```python
# Plot 3: Petal Length Distribution
plt.subplot(2, 3, 3)
sns.histplot(iris_df["PetalLengthCm"], kde=True, color='lightgreen')
plt.title("Petal Length Distribution")
```

```python
# Plot 4: Petal Width Distribution
plt.subplot(2, 3, 4)
sns.histplot(iris_df["PetalWidthCm"], kde=True, color='gold')
plt.title("Petal Width Distribution")
```

```python
# Plot 5: Boxplot for Sepal Length
plt.subplot(2, 3, 5)
sns.boxplot(y=iris_df["SepalLengthCm"], color='skyblue')
plt.title("Sepal Length Boxplot")
```

```python
# Plot 6: Boxplot for Sepal Width
plt.subplot(2, 3, 6)
sns.boxplot(y=iris_df["SepalWidthCm"], color='salmon')
plt.title("Sepal Width Boxplot")
```

```python
# Plot 7: Boxplot for Petal Length
plt.subplot(3, 4, 7)
sns.boxplot(y=iris_df["PetalLengthCm"], color='lightgreen')
plt.title("Petal Length Boxplot")
```

```python
# Plot 8: Boxplot for Petal Width
plt.subplot(3, 4, 8)
sns.boxplot(y=iris_df["PetalWidthCm"], color='gold')
plt.title("Petal Width Boxplot")
```

```python
# Plot 9: Swarmplot for Sepal Length
plt.subplot(3, 4, 9)
sns.swarmplot(y=iris_df["SepalLengthCm"], color='skyblue')
plt.title("Sepal Length Swarmplot")
```

```python
# Plot 10: Swarmplot for Sepal Width
plt.subplot(3, 4, 10)
sns.swarmplot(y=iris_df["SepalWidthCm"], color='salmon')
plt.title("Sepal Width Swarmplot")
```

```python
# Plot 11: Swarmplot for Petal Length
plt.subplot(3, 4, 11)
sns.swarmplot(y=iris_df["PetalLengthCm"], color='lightgreen')
plt.title("Petal Length Swarmplot")
```

```python
# Plot 12: Swarmplot for Petal Width
plt.subplot(3, 4, 12)
sns.swarmplot(y=iris_df["PetalWidthCm"], color='gold')
plt.title("Petal Width Swarmplot")


plt.tight_layout()
plt.show()
```
