```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have "iris.csv" downloaded
iris = pd.read_csv("/Iris.csv")

# Create histograms for each numerical feature
features = ["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]  # Adjust as needed
rows, cols = 2, 2  # Adjust for number of features (2x2 grid)
fig, axes = plt.subplots(rows, cols, figsize=(10, 10))  # Create a figure and subplots

for i, feature in enumerate(features):
    row, col = divmod(i, cols)  # Calculate subplot positions
    iris[feature].hist(bins=10, ax=axes[row, col])  # Create histogram with 10 bins
    axes[row, col].set_title(feature)  # Set subplot title
    axes[row, col].set_xlabel(feature)  # Set x-axis label
    axes[row, col].set_ylabel("Frequency")  # Set y-axis label

# Adjust layout (optional)
plt.tight_layout()

# Display the histograms
plt.show()
```
